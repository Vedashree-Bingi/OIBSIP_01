```python
#importing the necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading csv file

data = pd.read_csv('Iris.csv')
data
```

```python
#Dropping the unwanted columns

data = data.drop(["Id"],axis=1)
data
```

```python
# Counting the  different Species

data["Species"].value_counts()
```

```python
# Checking for null values

data.isnull().sum()
```

```python
#VISUALIZATION

species_count =  data.Species.value_counts()
label = data.Species.unique().tolist()

# Pie Chart
plt.pie(species_count, labels=label,  autopct='%1.1f%%')
plt.title("Species Counts", fontsize=18)
plt.show()
```

```python
#Count plot for the Iris Species

plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=data, palette='viridis')
plt.title('Count Plot of Different Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Creating a box plot for sepal length by species

plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=data, palette='viridis')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# Create a box plot for sepal width by species

plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='SepalWidthCm', data=data, palette='viridis')
plt.title('Box Plot of Sepal width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal width')
plt.show()
```

```python
# PetalLengthCm vs Species

plt.bar(data['Species'],data['PetalLengthCm'])
plt.title("Petal Length vs Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.grid(True)
plt.show()
```

```python
# Creating a violin plot for PetalWidthCm by species

plt.figure(figsize=(8, 6))
sns.violinplot(x='Species', y='PetalWidthCm', data=data, palette='viridis')
plt.title('Violin Plot of PetalWidthCm by Species')
plt.xlabel('Species')
plt.ylabel('PetalWidthCm')
plt.show()
```

```python
#Heatmap of correlation matrix

#Compute the correlation matrix
corr_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Heatmap of Feature Correlations')
plt.show()
```
